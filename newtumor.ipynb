```python
# %%
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))


# %%
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import cv2
from tqdm import tqdm

# %%
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale=1./255,
                                   shear_range=0.2,
                                   zoom_range=0.2,
                                   horizontal_flip=True)

training_set = train_datagen.flow_from_directory("D:\\Prayag Files\\TIET\\Extras\\hackathons\\analog\\brain_tumor_dataset",
                                                 target_size=(64, 64),
                                                 batch_size=32,
                                                 class_mode='binary')


# %%
import os
import matplotlib.pyplot as plt
import cv2

# Define the data directory and categories
DATADIR = 'D:\\Prayag Files\\TIET\\Extras\\hackathons\\analog\\brain_tumor_dataset'
CATEGORIES = {'no': 'Normal', 'yes': 'Tumor'}

# Iterate through each category
for key, value in CATEGORIES.items():
    path = os.path.join(DATADIR, key)  # Create path to the category folder
    
    x = 0  # Initialize image counter
    fig, axes = plt.subplots(1, 3, figsize=(20, 10))  # Create a subplot with 1 row and 3 columns
    
    # Iterate over each image in the category folder
    for img in os.listdir(path):
        img_array = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)  # Convert to array
        axes[x].imshow(img_array, cmap='gray')  # Display the image in grayscale
        axes[x].axis('off')  # Hide axis
        x += 1
        
        # Break the loop after displaying 3 images
        if x == 3:
            break
    
    # Set the title for the subplot
    plt.suptitle(value, fontsize=20)
    plt.tight_layout()
    plt.show()


# %%
IMG_SIZE = 200

new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
plt.imshow(new_array, cmap='gray')
plt.show()

# %%
len(training_set)

# %%
print(training_set.class_indices)

# %%
# Define a CNN model with multiple outputs (binary classification + regression)
def build_model():
    # Image input layer
    input_img = tf.keras.layers.Input(shape=(64, 64, 3))
    
    # Convolutional layers
    x = tf.keras.layers.Conv2D(32, (3, 3), activation='relu')(input_img)
    x = tf.keras.layers.MaxPooling2D(pool_size=(2, 2))(x)
    x = tf.keras.layers.Conv2D(32, (3, 3), activation='relu')(x)
    x = tf.keras.layers.MaxPooling2D(pool_size=(2, 2))(x)
    x = tf.keras.layers.Flatten()(x)
    
    # Fully connected layers for classification
    x_cls = tf.keras.layers.Dense(128, activation='relu')(x)
    output_cls = tf.keras.layers.Dense(1, activation='sigmoid', name='output_cls')(x_cls)
    
    # Fully connected layers for regression (age, location, size)
    x_reg = tf.keras.layers.Dense(64, activation='relu')(x)
    output_age = tf.keras.layers.Dense(1, name='output_age')(x_reg)  # Example regression output for age
    output_loc = tf.keras.layers.Dense(2, name='output_loc')(x_reg)  # Example regression output for location (x, y)
    output_size = tf.keras.layers.Dense(1, name='output_size')(x_reg)  # Example regression output for size
    
    # Define the model with multiple outputs
    model = tf.keras.models.Model(inputs=input_img, outputs=[output_cls, output_age, output_loc, output_size])
    
    # Compile the model with appropriate loss functions and metrics
    model.compile(optimizer='adam',
                  loss={'output_cls': 'binary_crossentropy',  # Binary cross-entropy for classification
                        'output_age': 'mean_squared_error',   # Example mean squared error for age
                        'output_loc': 'mean_squared_error',   # Example mean squared error for location
                        'output_size': 'mean_squared_error'}, # Example mean squared error for size
                  metrics={'output_cls': 'accuracy'})  # Accuracy for classification output
    
    return model

# Build the model
model = build_model()

# Train the model
model.fit(training_set, epochs=30)

# %%
print(model.summary())

# %%
# Example of predicting with the model for all images in the dataset
def predict_all_images(model, directory):
    predictions = {'Filename': [], 'Classification': [], 'Age': [], 'Location': [], 'Size': []}
    
    for class_label, class_name in CATEGORIES.items():
        path = os.path.join(directory, class_label)
        
        for img in os.listdir(path):
            img_path = os.path.join(path, img)
            img_array = cv2.imread(img_path)
            img_resized = cv2.resize(img_array, (64, 64))
            img_resized = np.expand_dims(img_resized, axis=0)
            img_rescaled = img_resized / 255.0
            
            # Predict with the model
            predictions_cls, predictions_age, predictions_loc, predictions_size = model.predict(img_rescaled)
            
            # Store predictions
            predictions['Filename'].append(img)
            predictions['Classification'].append('Tumor' if predictions_cls[0][0] >= 0.5 else 'Normal')
            predictions['Age'].append(predictions_age[0][0])
            predictions['Location'].append((predictions_loc[0][0], predictions_loc[0][1]))
            predictions['Size'].append(predictions_size[0][0])
    
    # Create a DataFrame from predictions
    predictions_df = pd.DataFrame(predictions)
    
    return predictions_df

# Predict for all images in the dataset
predictions_df = predict_all_images(model, DATADIR)
print(predictions_df.head())

# %%
import os
import random
import numpy as np
from keras.preprocessing import image
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report

# Function to predict an image without visualization
def predict_image(model, file_path):
    img = image.load_img(file_path, target_size=(64, 64))
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    result = model.predict(img_array, verbose=0)
    
    return result

# Function to visualize predictions for a subset of images with true labels
def visualize_predictions_with_labels(model, no_dir, yes_dir):
    no_files = random.sample(os.listdir(no_dir), 5)
    yes_files = random.sample(os.listdir(yes_dir), 5)
    
    for file_name in no_files:
        file_path = os.path.join(no_dir, file_name)
        predictions = predict_image(model, file_path)
        prediction_cls = 'Tumor' if predictions[0][0][0] >= 0.5 else 'Normal'
        prediction_age = predictions[1][0][0]  # Example for age prediction
        prediction_loc = predictions[2][0][0], predictions[2][0][1]  # Example for location prediction (x, y)
        prediction_size = predictions[3][0][0]  # Example for size prediction
        
        img = image.load_img(file_path, target_size=(64, 64))
        plt.imshow(img)
        plt.title(f'True Label: No | Prediction: {prediction_cls}\nAge: {prediction_age}, Location: {prediction_loc}, Size: {prediction_size}')
        plt.axis('off')
        plt.show()
    
    for file_name in yes_files:
        file_path = os.path.join(yes_dir, file_name)
        predictions = predict_image(model, file_path)
        prediction_cls = 'Tumor' if predictions[0][0][0] >= 0.5 else 'Normal'
        prediction_age = predictions[1][0][0]  # Example for age prediction
        prediction_loc = predictions[2][0][0], predictions[2][0][1]  # Example for location prediction (x, y)
        prediction_size = predictions[3][0][0]  # Example for size prediction
        
        img = image.load_img(file_path, target_size=(64, 64))
        plt.imshow(img)
        plt.title(f'True Label: Yes | Prediction: {prediction_cls}\nAge: {prediction_age}, Location: {prediction_loc}, Size: {prediction_size}')
        plt.axis('off')
        plt.show()

# Function to generate classification report using all files
def generate_full_classification_report(model, no_dir, yes_dir):
    predictions_cls = []
    predictions_age = []
    predictions_loc = []
    predictions_size = []
    true_labels = []
    
    for class_label, directory in [('no', no_dir), ('yes', yes_dir)]:
        for file_name in os.listdir(directory):
            file_path = os.path.join(directory, file_name)
            predictions = predict_image(model, file_path)
            predictions_cls.append(predictions[0][0][0])
            predictions_age.append(predictions[1][0][0])
            predictions_loc.append(predictions[2][0])
            predictions_size.append(predictions[3][0][0])
            true_labels.append(0 if class_label == 'no' else 1)
    
    predictions_cls = np.array(predictions_cls)
    true_labels = np.array(true_labels)
    predictions_cls = np.where(predictions_cls >= 0.5, 1, 0)
    
    print(classification_report(true_labels, predictions_cls))

# Example usage:
no_dir = r"D:\Prayag Files\TIET\Extras\hackathons\analog\brain_tumor_dataset\no"
yes_dir = r"D:\Prayag Files\TIET\Extras\hackathons\analog\brain_tumor_dataset\yes"

# Visualize predictions for 10 random images (5 from each folder) with true labels
visualize_predictions_with_labels(model, no_dir, yes_dir)

# Generate classification report using all files from both folders
generate_full_classification_report(model, no_dir, yes_dir)
```

```python
import numpy as np 
import pandas as pd 


import os



import pathlib
import shutil
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import itertools
from keras.utils import to_categorical, image_dataset_from_directory
from keras.models import Sequential
from keras import layers
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPool2D, GlobalAveragePooling2D
from keras.layers import MaxPooling2D, GlobalMaxPooling2D, Activation
from keras.optimizers import RMSprop
from keras import optimizers
from keras.callbacks import ReduceLROnPlateau, EarlyStopping
from keras.models import Model
from tensorflow.keras import applications, losses
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
colors_dark = ["#1F1F1F", "#313131", '#636363', '#AEAEAE', '#DADADA']
colors_red = ["#331313", "#582626", '#9E1717', '#D35151', '#E9B4B4']
colors_green = ['#01411C','#4B6F44','#4F7942','#74C365','#D0F0C0']
```

```python
import pathlib
from tensorflow.keras.preprocessing import image_dataset_from_directory


train_directory = "D:/Prayag Files/TIET/Extras/hackathons/analog/Research Papers/Mine/training"
test_directory = "D:/Prayag Files/TIET/Extras/hackathons/analog/Research Papers/Mine/testing"


train_path = pathlib.Path(train_directory)
test_path = pathlib.Path(test_directory)

train_df = image_dataset_from_directory(
    train_path,
    image_size=(128, 128),  
    validation_split=0.3,   
    subset="training",      
    shuffle=True,           
    batch_size=25,          
    seed=123                
)


validation_df = image_dataset_from_directory(
    train_path,
    image_size=(128, 128),  
    validation_split=0.3,   
    subset="validation",   
    shuffle=True,          
    batch_size=25,          
    seed=123                
)


test_df = image_dataset_from_directory(
    test_path,
    image_size=(128, 128), 
    shuffle=True,           
    batch_size=25,          
    seed=123               
)

print(f"Training batches: {len(train_df)}")
print(f"Validation batches: {len(validation_df)}")
print(f"Testing batches: {len(test_df)}")
```

```python
autotune = tf.data.AUTOTUNE
train_df = train_df.cache()
validation_df = validation_df.cache()
test_df = test_df.cache()

pf_train = train_df.prefetch(buffer_size = autotune)
pf_test = test_df.prefetch(buffer_size = autotune)
pf_val = validation_df.prefetch(buffer_size = autotune)
```

```python

directory = r"D:\Prayag Files\TIET\Extras\hackathons\analog\Research Papers\Mine\training"
path_for_training_data = pathlib.Path(directory)

train_df224 = image_dataset_from_directory(path_for_training_data,
                                           image_size=(224, 224),
                                           validation_split=0.3,
                                           subset="training",
                                           shuffle=True,
                                           batch_size=25,
                                           seed=123)

validation_df224 = image_dataset_from_directory(path_for_training_data,
                                                image_size=(224, 224),
                                                validation_split=0.3,
                                                subset="validation",
                                                shuffle=True,
                                                batch_size=25,
                                                seed=123)


directory = r"D:\Prayag Files\TIET\Extras\hackathons\analog\Research Papers\Mine\testing"
path_for_testing_data = pathlib.Path(directory)

test_df224 = image_dataset_from_directory(path_for_testing_data,
                                          image_size=(224, 224),
                                          shuffle=True,
                                          batch_size=25,
                                          seed=123)
```

```python
autotune = tf.data.AUTOTUNE
train_df224 = train_df224.cache()
validation_df224 = validation_df224.cache()
test_df224 = test_df224.cache()

pf_train224 = train_df224.prefetch(buffer_size = autotune)
pf_test224 = test_df224.prefetch(buffer_size = autotune)
pf_val224 = validation_df224.prefetch(buffer_size = autotune)
```

```python
data_augmentation = tf.keras.Sequential()
data_augmentation.add(layers.RandomRotation(0.3))
data_augmentation.add(layers.RandomFlip("horizontal_and_vertical"))
data_augmentation.add(layers.RandomZoom(0.1))
data_augmentation.add(layers.RandomContrast(0.2))
data_augmentation.add(layers.RandomBrightness(0.2))
```

```python
data_augmentation2 = tf.keras.Sequential()
data_augmentation2.add(layers.RandomRotation(0.3))
data_augmentation2.add(layers.RandomFlip("horizontal_and_vertical"))
data_augmentation2.add(layers.RandomZoom(0.1))
data_augmentation2.add(layers.RandomContrast(0.2))
data_augmentation2.add(layers.RandomBrightness(0.2))
```

```python

image_size=128
effnet0 = applications.EfficientNetB0(weights='imagenet',include_top=False,input_shape=(image_size,image_size,3))
```

```python
image_size224=224
effnet0_224 = applications.EfficientNetB0(weights='imagenet',include_top=False,input_shape=(image_size224,image_size224,3))
```

```python
modeleffnet0 = Sequential()
modeleffnet0.add(tf.keras.Input(shape=(128, 128, 3)))
modeleffnet0.add(data_augmentation)
modeleffnet0.add(effnet0)
modeleffnet0.add(layers.GlobalAveragePooling2D())
modeleffnet0.add(layers.Dropout(rate=0.5))
modeleffnet0.add(layers.Dense(4, activation='softmax'))
```

```python
modeleffnet0224 = Sequential()
modeleffnet0224.add(tf.keras.Input(shape=(224, 224, 3)))
modeleffnet0224.add(data_augmentation2)
modeleffnet0224.add(effnet0_224)
modeleffnet0224.add(layers.GlobalAveragePooling2D())
modeleffnet0224.add(layers.Dropout(rate=0.5))
modeleffnet0224.add(layers.Dense(4, activation='softmax'))
```

```python
modeleffnet0.compile(loss=losses.SparseCategoricalCrossentropy(),optimizer = 'Adam', metrics= ['accuracy'])
modeleffnet0224.compile(loss=losses.SparseCategoricalCrossentropy(),optimizer = 'Adam', metrics= ['accuracy'])
```

```python

reduce_lr = ReduceLROnPlateau(monitor = 'val_accuracy', factor = 0.3, patience = 2, min_delta = 0.001,
                              mode='auto',verbose=1)
```

```python

history = modeleffnet0.fit(pf_train,validation_data = (pf_val), epochs =12, verbose=1, batch_size=32,
                   callbacks=[reduce_lr])
```

```python
def matrix(model,test_images,test_labels):    
    pred_labels = np.argmax(model.predict(test_images), axis=1)
    conf_matrix = confusion_matrix(test_labels, pred_labels)

    accuracy = accuracy_score(test_labels, pred_labels)
    precision = precision_score(test_labels, pred_labels, average='weighted', zero_division=0)
    recall = recall_score(test_labels, pred_labels, average='weighted', zero_division=0)
    f1 = f1_score(test_labels, pred_labels, average='weighted', zero_division=0)

    plt.figure(figsize=(10, 8))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.title('Confusion Matrix')
    plt.show()
    
    print(f"Accuracy: {accuracy}")
    print(f"Precision: {precision}")
    print(f"Recall: {recall}")
    print(f"F1 Score: {f1}")

    
matrix(modelresnet50,test_images224,test_labels224)
```
